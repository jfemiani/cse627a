#### One-Hot Encoding
- Technique used to represent categorical variables as numerical values
- Each category is assigned a unique binary vector of length equal to the number of categories
- Only one element is set to 1 and the others are set to 0
- The position of the 1 in the vector corresponds to the index of the category in the original categorical variable


Suppose we have a target variable with three categories: "red", "green", and "blue". To perform one-hot encoding, we would assign each category a unique binary vector of length three:

- "red" -> [1 0 0]
- "green" -> [0 1 0]
- "blue" -> [0 0 1]

Then, for each data point with a particular category, we would use the corresponding binary vector as the label. For example, if a data point belongs to the "green" category, its label would be [0 1 0].


```python
import torch

# Define a tensor of categorical labels
labels = torch.tensor([0, 1, 2, 1, 0])

# Perform one-hot encoding
one_hot = torch.nn.functional.one_hot(labels)

print(one_hot)
```

#### Classification using $p(C_k|x)$
- Classification is the task of predicting a categorical label for a given input
- $p(C_k|x)$ is the posterior probability of class $C_k$ given input $x$
- $p(C_k|x)$ can be modeled using Bayes theorem, involving likelihood and priors 

$$p(C_k|x) = \frac{p(x|C_k)p(C_k)}{p(x)} = \frac{p(x|C_k)p(C_k)}{\sum_{j}p(x|C_j)p(C_j)}$$

where:

- $p(x|C_k)$ is the likelihood of the data given class $C_k$
- $p(C_k)$ is the prior probability of class $C_k$
- $p(x)$ is the marginal likelihood of the data
- $\sum_{j}p(x|C_j)p(C_j)$ is the normalizing constant or evidence, which ensures that the posterior probabilities sum up to 1.


#### Linear function $y(x) = f(w^T x + w_0)$
- $w$ is a weight vector and $w_0$ is a bias term
- $x$ is the input feature vector
- $f(\cdot)$ is a nonlinear activation function (e.g., sigmoid, softmax)
- $y(x)$ is the predicted output of the linear model

### 4.1. Discriminative Models
- Discriminative models directly model the decision boundary between classes.
- Logistic regression is a common discriminative model used for binary classification. It models the posterior probability of the positive class.
- Softmax regression is a generalization of logistic regression for multi-class classification.

This figure illustrates the geometry of a linear discriminant function in two dimensions.

<img src="Figure4.1.png" width="30%"/>


- The decision surface, shown in red, is **perpendicular to w**, and its displacement from the origin is controlled by the bias parameter w0.
- The signed orthogonal distance of a general point x from the decision surface is given by the projection of x onto w.
- The predicted class for a point x is determined by the sign of the discriminant function y(x) = w^T x + w0.

>  The $\mathbf{w}$ vector is not unique -- you can scale it by any nonzero factor as long as you scale $w_o$ as well.  If you normalize it, then $\mathbf{w}$ is a _normal vector_ of the separating hyperplane.

#### Least Squares for Classification
- Not really used.... (!)
- Least squares approach finds the line that minimizes the sum of squared distances between the data points and the line
- It assumes the class-conditional densities are Gaussian with equal covariance matrices
- Assumes the classes are equiprobable
- The resulting decision boundary is a linear function of the input variables
- Least squares is highly sensitive to outliers and can lead to overfitting

- The idea is to project all of your data onto a single line, and then find a 1D threshold
- In 2+ dimensions, there are many lines that could be used, but depending on the line, the ability to separate the projected points is different
- One option could be to use a line that connects the means of the two classes, but this does not take into account the shape of the distributions. The covariance also needs to be considered.
- Fisher finds a direction where the within-class variance of the points **after projection** is as small as possible.
   - Solving this involves finding an eigenvector of the sum of the within-class covariances. 
- Then there are several ways to find a 1D threshold.
  - If the variances are the same, then the optimal threshold is $y_0 = \frac{1}{2}(m_1 + m_2)$

### 4.2. Probabilistic Generative Models
- Generative models model the joint probability of the input and output. They can be used to generate new data.

- The assumption of Gaussian class-conditional densities is reasonable for many problems and allows for analytical solutions
- The maximum likelihood solution for parameters of a Gaussian distribution can be obtained by solving the normal equations
- The discriminant functions for two-class Gaussian distributions are quadratic functions of x
- The decision boundary is linear in the case of two Gaussian distributions with identical covariance matrices

<img src="Figure4.11a.png" width="30%" /><img src="Figure4.11b.png" width="30%"/>


- The sigmoid function can be derived from the Gaussian discriminant functions
    - Let the *log odds ratio* be $$a = \ln \frac{p(x|C_1)p(C_1)}{p(x|C_2)p(C_2)} =  \ln \frac{p(x, C_1)}{p(x, C_2)}$$.
    - Then the postetior $$
    \begin{align}
    p(C_1|x) &= \frac{p(x, C_1)}{p(x, C_1) + p(x, C_2)} & (\text{Bayes Thm}) \\
             &= \frac{1}{1 + p(x, C_2)/p(x, C_1)}   & (\text{Divide by $p(x, C_1)$}) \\
             &= \frac{1}{1+\exp{(-a)}}  & (\text{Defn of $a$}) \\
             &= \sigma(a) & (\text{Defined now!})
    \end{align}$$
    - This is the _sigmoid_ function for binary classification. 
    - The probability of class $C_2= 1-\sigma(a)$.
    - The derivative is $\sigma(a)(1-\sigma(a))$

```python
%run plot-sigmoid.py
```

- The softmax function is used for multi-class classification and outputs class probabilities that sum to 1
    - Note that $a$ is a log odds **ratio**  between two classed.
    - For multiple classes, **define** a_k to be $$a_k =\ln p(x, C_k)$$  (so $a_k$ is not a ratio)
    - Then $$
    \begin{align}
    p(C_k|x) &= \frac{\exp(a_k)}{\sum_j \exp(a_j)} & (\text{Bayes Thm & the $\ln$ in the defn of $a_k$})\\
             &= \boldsymbol{\sigma}(\mathbf{a})  & (\text{Defined here!})
    \end{align}$$

- The $a_k$ are often called **logits**
- Linear regression on the logits is logistic regression, which is classification!

#### 4.3.2 Logistic Regression

In logistic regression, we assume that the probability of a data point belonging to class 1 (vs. class 0) is given by the sigmoid function:

$$ p(C_1|\phi_n) = y_n = \sigma(w^T \phi_n) $$

where $\phi_n$ is the feature vector for the $n$th data point, $w$ is the weight vector, and $\sigma(\cdot)$ is the sigmoid function.

Given a set of training data $\{(\phi_n, t_n)\}$, where $t_n \in \{0,1\}$ is the target variable for the $n$th data point, we can write the likelihood function as:

$$ p(\mathbf{t}|\mathbf{w}) = \prod_{n=1}^N y_n^{t_n} (1 - y_n)^{1-t_n} $$

where $\mathbf{t} = (t_1, \dots, t_N)^T$ and $\mathbf{y} = (y_1, \dots, y_N)^T$.



The log-likelihood function is then:

$$ \ln p(\mathbf{t}|\mathbf{w}) = \sum_{n=1}^N \left( t_n \ln y_n + (1-t_n) \ln(1-y_n) \right) $$

We can use gradient descent to find the maximum of the log-likelihood function and obtain the optimal weight vector $\mathbf{w}$. This involves taking the gradient of the log-likelihood with respect to $\mathbf{w}$:

$$ \nabla \ln p(\mathbf{t}|\mathbf{w}) = \sum_{n=1}^N (\mathbf{y}_n - \mathbf{t}_n) \phi_n $$

where $\mathbf{y}_n = (y_n, 1-y_n)^T$ and $\phi_n = (\phi_n, 1)^T$ is the augmented feature vector.

Note that the likelihood function for logistic regression is a product of Bernoulli distributions, one for each data point. This makes it different from the likelihood function for Gaussian discriminant analysis, which is a product of multivariate Gaussian distributions. 


#### 4.3.3 Iterative Reweighted Least Squares
- Logistic regression does not have a closed-form solution (unlike linear regression)
- We must use an iterative approach (like gradient descent)
- Newton-Rhapson iteration is gradient-descent on steriods
   - Instead of just taking the gradient (and $M$-vector), we take the second-derivatives
   - The second derivatives are an $M\times M$ matrix of mixed partial derivatives called the Hessian matrix $$H_{ij} = \frac{\partial^2 E}{\partial a_i \partial a_j}$$
   - The first and second derivatives form quadratic local approximation to the function (Taylor approximation)
   - It is bowl-shaped, so unlike gradient descent you can solve for a minimum
   - Of course the quadratic was only an approximation to the error, but it gets $w$ closer (usually)
   - This takes fewer (but more expensive) iterations than gradient descent. 
- You can **decompose** the hessian matrix in a useful way
  $$H = \Phi^T R \Phi$$
  where 
  $$R = \operatorname{diag}\left[\  y_n (1-y_n) \  \right]$$ 
  and the $y_n$ are based on the current value of $\mathbf{w}^{(k)},$
- At each step $R_{nn}$ can be interpeted as *weights* applied to the data points
   - TANGENT: maybe $\sqrt{R_{nn}}$ would be better?? Then $H = (\sqrt{R}\Phi)^2$

```python
import numpy as np
np.seterr(over='ignore')  # np.exp does not like large numbers

def sigmoid(x):
    """Sigmoid function."""
    return 1 / (1 + np.exp(-x))

    
def irwls(X, y, max_iter=100, tol=1e-6, alpha=1e-4):
    """Iterative reweighted least squares for logistic regression."""
    # Initialize weights and predictions
    w = np.zeros(X.shape[1])
    y_pred = sigmoid(X @ w)
    
    for i in range(max_iter):
        # Compute the diagonal weighting matrix R
        R = np.diag(y_pred * (1 - y_pred))
        
        # Compute the gradient and Hessian with regularization
        grad = X.T @ (y - y_pred)
        H = X.T @ R @ X
        H = H + alpha * np.eye(X.shape[1])  # Regularize!
        
        # Compute the Newton-Raphson update
        w_new = w - np.linalg.inv(H) @ grad
        
        # Compute the new predictions and check convergence
        y_pred_new = sigmoid(X @ w_new)
        if np.allclose(w_new, w, atol=tol) or np.allclose(y_pred_new, y_pred, atol=tol):
            break
        
        # Update the weights and predictions
        w = w_new
        y_pred = y_pred_new
    
    return w
```

### 4.4. The Laplace Approximation
- The MLE estimates for $w$ can overfit (we saw that with $w \rightarrow \infty$!)
- The max posterior (after regularizing $w$) can help with that. 
- For Bayesian estimation, we need to model the posterior $p(w|X, y)$ 
- It is not Gaussian, and it is hard to model!!
- The Laplace approximation is a technique for approximating the posterior d $p(w|X, y)$.
- It is useful when the posterior distribution is difficult to compute exactly.
    - We find a mode (max posterior) estimate 
    - Then we compute a local quadratic approximation (Hessian) of the log posterior
    - We use the Gaussian approximation for Bayesian Inference
    

### 4.5. Bayesian Logistic Regression
- Bayesian logistic regression is a probabilistic approach to logistic regression.
- It involves placing a prior distribution on the model parameters and computing the posterior distribution using Bayes' theorem.
- Bayesian logistic regression allows for uncertainty estimates in the model parameters and predictions.
